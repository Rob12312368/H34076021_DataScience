```python
import pandas as pd
import numpy as np
data=pd.read_csv('IMDB-Movie-Data.csv')
data.head()
```

```python
df=data.copy()
df=df[df['Year']==2016].sort_values('Rating', ascending=False) #choose the movies with highest rating in 2016
df.head(3)
```

```python
actors=data.copy()
actors['Actors']=actors['Actors'].str.replace(r'\|\s*','|',regex=True).str.split('|')# convert string into list
actors=actors.explode('Actors') #split the list into multiple rows


highest_revenue=actors.copy()
highest_revenue=actors.groupby('Actors').mean().sort_values('Revenue (Millions)',ascending=False) #make actors into groups and sort

highest_revenue=highest_revenue[highest_revenue['Revenue (Millions)']==highest_revenue['Revenue (Millions)'].max()]# choose the rows with the gihest revenue

highest_revenue
    
```

```python
emma=actors.copy()
emma=emma.groupby('Actors')
emma=emma.get_group('Emma Watson')
pd.DataFrame({'Average Rating':emma['Rating'].mean()}, index=['Emma Watson'])
```

```python
directors=actors.copy()
directors=directors.drop_duplicates(subset=['Director', 'Actors'],keep='last') #avoid counting data with same director and actor
directors=directors.groupby('Director').size().sort_values(ascending=False)
directors=directors.nlargest(3,keep='all') #top3 and keep the data with the same value
pd.DataFrame(directors,columns=['Count of Collaboration'])
```

```python
genres=actors.copy()
genres['Genre']=genres['Genre'].str.split('|')
genres=genres.explode('Genre') #duplicate data 
genres=genres.drop_duplicates(subset=['Genre', 'Actors'],keep='last')
genres=genres.groupby('Actors')
genres=genres.size().sort_values(ascending=False)
genres=genres.nlargest(2,keep='all')#top2 and keep the data with the same value
pd.DataFrame(genres,columns=['Count of Genres'])
```

```python
gap=actors.copy()
gap=gap.groupby('Actors')
result=gap['Year'].max()-gap['Year'].min() #count the gap
result=result.sort_values(ascending=False) 
result=result.nlargest(3,keep='all') #top3 and keep the data with the same value
result=pd.DataFrame(result)
result
```

```python
cooperate=data.copy()
actors=cooperate['Actors'] #actors column
myseries=cooperate[cooperate['Actors'].str.find('Johnny Depp')!=-1] #find data with johnny depp
myseries=myseries['Actors'].str.split('|')
myseries=myseries.explode()
myseries=set(map(lambda x:x.strip(),myseries))#prevent ' Johnny Depp'
result=myseries.copy() #direct collaborate
former=0
now=0
while(1):
    former=len(result)
    for index, value in actors.items():
        for actor in result:
            if actor in value: #if the actor in result is in current row, then add into result
                temp=value.split('|')
                temp=set(map(lambda x:x.strip(), temp))
                result=result.union(temp)
                now=len(result) 
    if now==former: #if no more data
        break
result.remove('Johnny Depp')
pd.DataFrame(result,columns=['Collaborate Actors'])
```
